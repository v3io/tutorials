```python
HOROVOD_JOB_NAME = "horovod-tf-benchmark"
```

```python
from v3io_gputils.mpijob import MpiJob

job = MpiJob(HOROVOD_JOB_NAME, 'iguaziodocker/horovod:0.1.1', ['scripts/tf_cnn_benchmarks/tf_cnn_benchmarks.py',
                                                        '--batch_size=32',
                                                        '--model=resnet50',
                                                        '--variable_update=horovod',
                                                        '--use_fp16=true',
                                                        '--xla=true'])
job.replicas(8).gpus(1)
job.submit()
```
